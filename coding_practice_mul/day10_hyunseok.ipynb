```python
1.
리스트에 있는 숫자들의 최빈값(mode)을 구하는 프로그램(함수)을 만들어라.
[12, 17, 19, 17, 23] = 17
[26, 37, 26, 37, 91] = 26, 37
[28, 30, 32, 34, 144] = 없다

최빈값 : 자료의 값 중에서 가장 많이 나타난 값 
① 자료의 값이 모두 같거나 모두 다르면 최빈값은 없다.
```

```python
def mode(x):
    if len(x)==len(set(x)):
        return '없다'
    else:
        count=0
        b=0
        c=0
        for i in x:
            if count<x.count(i):
                count=x.count(i)
                b=i
            elif count==x.count(i):
                c=i

    if c !=0:
        return b, c
    else:
        return b

list1=[12, 17, 19, 17, 23]
list2=[26, 37, 26, 37, 91]
list3=[28, 30, 32, 34, 144]

print(mode(list1))
print(mode(list2))
print(mode(list3))
        
    

```

```python
4. 리스트에 있는 숫자들의 중앙값(중위수, median)을 구하는 프로그램을 만들어라.

[7, 9, 14] = 9
[24, 31, 35, 49] = 33
[17, 37, 37, 47, 57] = 37

중앙값 : 자료를 작은 값에서부터 크기순으로 나열할 때 중앙에 위치한 값
① 자료의 개수가 홀수이면 가운데 위치한 값이 중앙값이다.
② 자료의 개수가 짝수이면 가운데 위치한 두 값의 평균이 중앙값이다.
```

```python
list1=[7, 9, 14]
list2=[24, 31, 35, 49]
list3=[17, 37, 37, 47, 57]

def median(x):
    if len(x)%2 !=0:
        return x[round(len(x)/2)-1]
    else:
        return (x[len(x)//2-1]+x[len(x)//2])/2

print(median(list1))
print(median(list2))
print(median(list3))
```

```python
6.
아래와 같은 결과를 출력하는 function을 구현하라
OneEditApart("cat", "dog") = false
OneEditApart("cat", "cats") = true 
OneEditApart("cat", "cut") = true 
OneEditApart("cat", "cast") = true 
OneEditApart("cat", "at") = true
OneEditApart("cat", "acts") = false 
한 개의 문자를 삽입, 제거, 변환 했을때 두 문자열이 동일한지를 판별하는 OneEditApart 함수를 작성하시오.
```

```python
def OneEditApart(str1, str2):
    if len(str1)<len(str2):
        if set(str1).issubset(set(str2)):
            return True
    elif len(str1)>len(str2):
        if set(str2).issubset(set(str1)):
            return True
    elif len(str1)==len(str2):
        if len(str1)==len(set.intersection(set(str1),set(str2)))+1:
            return True
        else:
            return False

        
    




print(OneEditApart("cat", "dog"))
print(OneEditApart("cat", "cats"))
print(OneEditApart("cat", "cut"))
print(OneEditApart("cat", "cast")) 
print(OneEditApart("cat", "at"))
print(OneEditApart("cat", "acts")) # 이건 구현 못함
```

```python
<어려움>
9. 당신은 A 인터넷 카페 운영자이다.
당신의 인터넷 카페에는 휴대폰 번호 게시가 금지되어 있다.
하지만 일부 회원들이 편법을 동원하여 휴대폰 번호를 게시 후 불법 거래를 시도한다.
이를 체크하여 자동 삭제를 할 수 있도록 휴대폰 번호 검사 알고리즘을 작성하시오.
(011~019 는 10자리 또는 11자리 휴대폰 번호이다. 010은 반드시 11자리여야만 한다.)

Input

영일영-구구칠8-일구팔사
0일영.칠칠칠팔.이삼사
영 일 칠 삼 칠 오 팔 이 팔 이
영일일 34구구 4 오 9 이

Output

01099781984 true
0107778234 false
0173758282 true
01134994592 true
```

```python
import re

def inspection(x):
    x=x.replace('영', '0')
    x=x.replace('일', '1')
    x=x.replace('이', '2')
    x=x.replace('삼', '3')
    x=x.replace('사', '4')
    x=x.replace('오', '5')
    x=x.replace('육', '6')
    x=x.replace('칠', '7')
    x=x.replace('팔', '8')
    x=x.replace('구', '9')
    x=x.replace('-', '')
    x=x.replace(' ', '')
    x=x.replace('.', '')

    if re.match('010[0-9]{8}', x):
        print (x, ': True')
    elif re.match('01[1-9]{7,8}', x):
        print(x, ': True')
    else:
        print(x, ': False')

a='영일영-구구칠8-일구팔사'
b='0일영.칠칠칠팔.이삼사'
c='영 일 칠 삼 칠 오 팔 이 팔 이'
d='영일일 34구구 4 오 9 이'

print(inspection(a))
print(inspection(b))
print(inspection(c))
print(inspection(d))
```

```python
<어려움>
10. 카프리카 수
카프리카 수란 인도의 수학자 D.R.카프리카의 의해 정의한 수이다.
어떤 수의 제곱수를 두 부분으로 나누어 더하였을 때 다시 원래의 수가 되는 수들을 의미한다.
예를 들어 45는 카프리카 수인데, 45² = 2025이고, 20+25 = 45이기 때문이다.

어떤 수를 입력 받고 그 수가 카프리카 수인지 아닌지를 출력하는 함수를 만드시오.

입력 (유형 1)

print(kaprekar_number(1))  # 1² =1, 1 = 1
print(kaprekar_number(45))
print(kaprekar_number(297))   #297의 제곱 = 88209 => 88+209 = 297
print(kaprekar_number(3213))
출력 (유형 1)

True
True
True
False
```

```python
def kaprekar_number(x):
    square=str(x**2)

    if x<=3:
        if x==1:
            return True
        else:
            return False
    elif len(square)%2 == 0:
        num=int(square[:len(square)//2])+int(square[len(square)//2:])
        if x==num:
            return True
        else:
            return False
    elif len(square)%2 !=0:
        num1=int(square[:len(square)//2])+int(square[len(square)//2:])
        num2=int(square[:1+len(square)//2])+int(square[1+len(square)//2:])
        if x==num1 or x==num2:
            return True
        else:
            return False

print(kaprekar_number(1))  # 1² =1, 1 = 1
print(kaprekar_number(45))
print(kaprekar_number(297))   #297의 제곱 = 88209 => 88+209 = 297
print(kaprekar_number(3213))
```
