```python
import numpy as np
import pandas as pd
```

```python
samsung = pd.read_csv('E:/Downloads/005930.KS.csv') ; samsung
```

```python
ohlcvs= samsung.columns[1:]

for ohlcv in list(ohlcvs):
    print(f'{ohlcv}: ')
    print(np.array(samsung[ohlcv]).min())
    print(np.array(samsung[ohlcv]).max())
    print(np.array(samsung[ohlcv]).mean())
    print(np.array(samsung[ohlcv]).std())
    print(np.array(samsung[ohlcv]).var())
    print()
```

```python
def Zscores(x_series):
    return (x_series - x_series.mean()) / x_series.std()
```

```python
x = pd.DataFrame()

for ohlcv in ohlcvs:
    series1 = Zscores(samsung[ohlcv])
    x[ohlcv]=series1
```

```python
x= x.set_index(samsung['Date']) #표준화
x
```

```python
today_close = x['Close'][-1]
```

```python
y=abs(x['Close'] - today_close)[:-1]
x['similiar']=y
```

```python
x.sort_values(by='similiar')
```
