```python
'''
유클리드 호제법(최대공약수/최소공배수)
1) 뺄셈 2) 나눗셈

1) 뺄셈 
8, 6 => 서로 빼서 큰거에서 뺀다
2, 6
2, 4
2, 2 => 값은 값 2

2) 나눗셈
8, 6 => 몫:1, 나머지:2
나머지가 0이 아니면 -> 젯수를 피젯수에게, 나머지를 젯수에게
6, 2 => 몫:3, 나머지:0
나머지가 0일때 -> 젯수가 최대공약수가 됨.


ex) 8과 6의 최대공약수?
약수 :
8: 1,2,4,8
6: 1,2,3,6

8,6의 공약수 : 1,2 
      최대공약수 : 2

8:8,16,24,..
6:6,12,18,24,...
8과 6의 공배수 : 24, 48,...
        최소공배수 : 24

24(최소공배수)= 8*6 / 2(최대공약수)

'''
```

```python
def Euclidean_algorithm_GCD(m,n):
    i=0
    while m != n:
        if m>n:
            m = m - n
        elif m<n:
            n = n - m
        i+=1
    return m
        
def Euclidean_algorithm_LCM(m,n):
    return m*n // Euclidean_algorithm_GCD(m,n)
```

```python
num1=int(input('숫자 입력'))
num2=int(input('숫자 입력'))

print(f'{num1}과 {num2}의 최대공약수 : {Euclidean_algorithm_GCD(num1, num2)}')
print(f'{num1}과 {num2}의 최대공배수 : {Euclidean_algorithm_LCM(num1, num2)}')
```

```python
# 2.
#  1~1000에서 각 숫자의 개수 구하기

# 예로 10 ~ 15 까지의 각 숫자의 개수를 구해보자

# 10 = 1, 0
# 11 = 1, 1
# 12 = 1, 2
# 13 = 1, 3
# 14 = 1, 4
# 15 = 1, 5

# 그러므로 이 경우의 답은 0:1개, 1:7개, 2:1개, 3:1개, 4:1개, 5:1개
```

```python
numList=[]
zero=0
one=0
two=0
three=0
four=0
five=0
six=0
seven=0
eight=0
nine=0


for num in range(1,1001):
    listlization=list(str(num))
    for i in range(len(listlization)):
        numList.append(listlization[i])

for i in range(len(numList)):
    if numList[i]=='0':
        zero+=1
    elif numList[i]=='1':
        one+=1
    elif numList[i]=='2':
        two+=1
    elif numList[i]=='3':
        three+=1
    elif numList[i]=='4':
        four+=1
    elif numList[i]=='5':
        five+=1
    elif numList[i]=='6':
        six+=1
    elif numList[i]=='7':
        seven+=1
    elif numList[i]=='8':
        eight+=1
    elif numList[i]=='9':
        nine+=1

print(f'0:{zero}, 1:{one}, 2:{two}, 3:{three}, 4:{four}, 5:{five}, 6:{six}, 7:{seven}, 8:{eight}, 9:{nine}')

```

```python
# 3.
# 시저 암호는, 고대 로마의 황제 줄리어스 시저가 만들어 낸 암호인데, 
# 예를 들어 알파벳 A를 입력했을 때, 그 알파벳의 n개 뒤에 오는
# (여기서는 예를 들 때 3으로 지정하였다)알파벳이 출력되는 것이다. 
# 예를 들어 바꾸려는 단어가 'CAT"고, n을 5로 지정하였을 때 "HFY"가 되는 것이다.

# 어떠한 암호를 만들 문장과 n을 입력했을 때 암호를 만들어 출력하는 프로그램을 
# 작성해라.
```

```python
def convert_to_alpha(words, n):
    if n<26:
        alphaList = list('ABCDEFGHIJKLMNOPQRSTUVWXYZ')*2
    else:
        alphaList = list('ABCDEFGHIJKLMNOPQRSTUVWXYZ')*round(n/10)
    temp_words=[]
    
    for alpha in str(words):
        temp_words.append(alphaList[alphaList.index(alpha)+n])

    return ''.join(temp_words)
```

```python
convert_to_alpha('CAT', 5)
```

```python
convert_to_alpha('HYUNSEOK', 30)
```
