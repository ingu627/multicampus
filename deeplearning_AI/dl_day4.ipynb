```python
#pca=PCA() #n_components를 생략하면 기본 차원이 됨
pca=PCA(n_components=3)
pca.fit(x_scaled)
x_pca=pca.transform(x_scaled)
```

```python
#print(x_pca)
#0	5.1	3.5	1.4	0.2 -> 표준화
#[-9.00681170e-01  1.01900435e+00 -1.34022653e+00 -1.31544430e+00] ->
#-> PCA(old 4 dim->new 4 dim)

pca.components_
```

```python
# 3차원 : pca1 ~ pca3
pca.explained_variance_
# 각 축이 설명하는 데이터의 표현력, 값이 클수록 설명을 많이 하고 있는 축
# 첫번째 축은 약 3을 갖고, 세번째 축 0.14이므로 
# 첫번째 pca1축이 세번째 pca3축보다 설명력이 약 20배 이상이다.

# 원래 데이터 
# x:신장, y:몸무게, z:머리카락길이

# array([2.93808505, 0.9201649 , 0.14774182])
# ex) 남녀를 식별(3가지 특성)
# 외형(PCA1) : 2.93808505 (73%)
# 목소리(PCA2) : 0.9201649 (22.8%)
# 이름(PCA3) : 0.14774182 (3%)
```

```python
ratio=pca.explained_variance_ratio_
# array([0.72962445, 0.22850762, 0.03668922])
```

```python
dfpca=pd.DataFrame(ratio, index=['pc1', 'pc2', 'pc3'], columns=['v'])
dfpca.plot.pie(y='v')
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix
```

```python
model = LogisticRegression(multi_class='multinomial')
```

```python
model.fit(x,y)
```

```python
pred=model.predict(x)
```

```python
#pca를 하지 않고, 원본 데이터 입력변수 4개를 사용한 모델의 정확도
#4차원 데이터 모델링
confusion_matrix(y, pred) #정확도 : 146 / 150
```

```python
# feature selection(2개 특성만 사용)
model2 = LogisticRegression(multi_class='multinomial')
model2.fit(x[:, :2], y)
```

```python
pred=model2.predict(x[:,:2])
```

```python
confusion_matrix(y, pred) #정확도 : 123  / 150
```

```python
pca=PCA()
pca.fit(x_scaled)
x_pca=pca.transform(x_scaled)
```

```python
model3=LogisticRegression(multi_class='multinomial')
model3.fit(x_pca, y)
```

```python
pred=model3.predict(x_pca)
confusion_matrix(y, pred)
```

```python
pred=model4.predict(x_pca[:,:2])
confusion_matrix(y, pred) #정확도 : 140  / 150
```
