```python
# 130개를 하나로 뭉쳐야 한다.
# 1880년~20101년까지 데이터 파일 읽어들이기
# 131개의 데이터프레임을 담기 위한 변수
pieces=[] #리스트에 데이터프레임을 요소로 저장

#feature engineering(특성 공학) : 데이터의 기존 변수로부터 연산을 통해
# 새로운 변수를 생성하는 작업 ex)국어, 영어 => 평균

for year in range(1880, 2011):
    df=pd.read_csv(f'e:/data/yob/yob{year}.txt',
        header=None, names=['name', 'sex', 'births'])
    df['year'] = year
    pieces.append(df)
```

```python
# len(pieces) # 리스트 안에 데이터프레임 131개가 요소로 저장

# 리스트내의 데이터프레임 모두 합쳐서 1개의 데이터프레임으로 만들자
names=pd.concat(pieces, ignore_index=True) #concatenate
names
```

```python
# names.to_csv('names.csv')
```

```python
# 성별에 따른 births의 합계
# 과연 1980~2010년까지 태어난 아이들의 수를 성별에 따라 비교
# ~에 따라 : groupby
names.groupby('sex').sum()

# 과연 1980~2010년까지 태어난 아이들의 수를 연도에 따라 성별에 따라 비교

```

```python
import matplotlib.pyplot as plt
```

```python
# names.groupby('sex').groupby('year').births.sum() 에러발생

# 성별에 따른 년도별 births의 합계 (2*131개 그룹)
# 두 개 이상의 컬럼을 기준으로 그룹화
names.groupby(['year', 'sex']).births.sum()

#pivot_table함수를 이용한 연산
# pivot : 중심축을 이용한다.
names.pivot_table('births', index='year', columns='sex', aggfunc=sum)
# aggfunc = agggregation function : 집계함수
# aggfunc = np.mean값이 default값
# 데이터프레임.pivot_table(연산에 사용될 데이터, index=행, columns=열, aggfunc=연산함수)

total_births=names.pivot_table('births', index='year', columns='sex', aggfunc=max)
```

```python
total_births
# 시각화 방법
# matplotlib, seaborn, 데이터프레임, plotly. 타블로,...

total_births.plot(title='Total births by sex and year')
```

```python
def add_prop(gr):
    gr['prop']=gr.births/gr.births.sum()
    return gr


# names.groupby(['year', 'sex']).대상열.연산함수()
names = names.groupby(['year', 'sex']).apply(add_prop)
 #데이터프레임.groupby().apply(함수) => 이 그룹에(262개 각각) 대해 함수가 적용
```

```python
names.groupby(['year','sex']).prop.sum()

names.groupby(['year','sex'])['prop'].sum()
# 연도별/성별 각 그룹별 prop값을 합을 출력

names.pivot_table('prop', index='year', columns='sex', aggfunc=sum) 
```

```python
# 각 연도/성별에 따라 그룹화 -> births 기준 내림차순 정렬
# -> 각 그룹별 상위 1000개의 데이터만 추출
```

```python
def getTop1000(gr):
    return gr.sort_values('births', ascending=False)[:1000]
    
top1000 = names.groupby(['year', 'sex']).apply(getTop1000)
```
