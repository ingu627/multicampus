```python
# sin, cos함수 그래프 함께 시각화
# 범위 : -pi ~ pi
# np.sin(-np.pi/2)

x=np.linspace(-np.pi, np.pi, 256)
ysin=np.sin(x)
ycos=np.cos(x)

plt.plot(x,ysin,'r--', x, ycos, 'g:')
plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi])
plt.yticks([-1,0,1])
```

```python
x1=np.linspace(0,5)
x2=np.linspace(0,2)
y1=np.cos(2*np.pi*x1)
y2=np.cos(2*np.pi*x1)*np.exp(-x1)

axes1=plt.subplot(2,1,1) #(줄,칸, 위치)
plt.plot(x1,y1,'yo-')

axes2=plt.subplot(2,1,2)
plt.plot(x2,y2,'r.-')
```

```python
# twinx : x축을 공유하는 axes를 생성
fig, ax=plt.subplots() # (6,4) 한 픽셀은 72
print(fig)
print(ax) #ax를 이용하여 특정 축을 공유 
# ex) x축은 공유, y축은 비공유
twinax=ax.twinx()
ax.plot([10,5,2,8,7], 'r-', label='y0')
ax.set_ylabel('y0')

twinax.plot([100,500,200,800,700], 'g:', label='y1')
twinax.set_ylabel('y1')
plt.show()
```
