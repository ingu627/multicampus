```python
# seaborn.pydata.org 참조
import seaborn as sns
```

```python
iris = sns.load_dataset('iris')
titanic = sns.load_dataset('titanic')
tips = sns.load_dataset('tips')
flights = sns.load_dataset('flights')
```

```python
flights
```

```python
# year별 승객수
flights.groupby('year').passengers.sum()
```

```python
sns.rugplot(x)
```

```python
# 커널 밀도 그래프 (곡선)
sns.kdeplot(x)
```

```python
sns.distplot(x)
```

```python
sns.distplot(x, rug=True)
```

```python
# countplot : 종류별로 데이터가 얼마나 있는지 출력
sns.countplot(x='class', data=titanic) # x : 종류 y : 빈도 
```

```python
sns.countplot(x='day', data=tips)
```

```python
# 다차원 데이터 시각화 
# 스캐터플랏(jointplot) : 2차원, 실수형
sns.jointplot(x='sepal_length' ,y='sepal_width' , data=iris)
```

```python
sns.jointplot(x='sepal_length' ,y='sepal_width' , data=iris, kind='kde')
```

```python
#pairplot : 다차원 데이터 시각화 (실수형 데이터는 scatter plot)
sns.pairplot(iris, hue='species') # hue는 범주형
```

```python
iris.corr()
```

```python
# 히트맵(열지도) : 2차원 데이터, 카테고리 형태 
```

```python
# data=titanic[['survived', 'pclass']]
# data

# 히트맵을 그리기 위해 데이터프레임 재구성 
data=titanic.pivot_table(index='survived', columns='pclass', aggfunc='size')
```

```python
sns.heatmap(data, cmap=sns.light_palette('yellow'), annot=True, fmt='d')
# cmap = 색상
# annot = 수치 나타내기 
# fmt = formatting / 'd' = 숫자로 나타내기
```

```python
# 실수, 종류가 섞여있는 데이터
# barplot, boxplot, pointplot, violinplot, stripplot, swarmplot
```

```python
# 요일별로 total_bill의 평균 
# 요일별로 tip의 합계 
tips.groupby('day').agg({'total_bill':'mean', 'tip':'sum'})
```

```python
sns.barplot(x='day', y='total_bill', data=tips)
```
