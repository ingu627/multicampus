```python
# 왜도(skew) : 어느 정도로 비대칭인지?
# - 음수인 경우 : 왼쪽이 긴 꼬리를 갖는 형태
# - 양수인 경우 : 오른쪽이 긴 꼬리를 갖는 형태
# - 0인 경우 : 평균과 중앙값이 같은 경우(완전 대칭)
# -0.5~0.5 : 대체적으로 대칭적이다.
# -1~-0.5, 0.5~1 : 어느 정도 치우쳐 있다.
# -1보다 작거나 1보다 큰 경우 : 많이 치우쳐 있다.
```

```python
df.skew()
```

```python
#첨도 : 확률분포의 뾰족한 정도, 특이값
# == 3 : 표준정규분포 (일때 mean=median=mode)

# 첨도가 3보다 작다 => 특이값이 거의 없다
# 첨도가 3보다 크다 => 특이값이 있다.
```

```python
print('% of women survived : ', 
train[train.Sex == 'female'].Survived.sum() / train[train.Sex == 'female'].Survived.count())
print('% of men survived : ', 
train[train.Sex == 'male'].Survived.sum() / train[train.Sex == 'male'].Survived.count())
```

```python
# (1, 2) : 한줄에 두개 칸
fig, ax =plt.subplots(1, 2, figsize=(16,8))
# fig = figure에 대한 객체 정보
# ax = 축
# subplots를 하면 fig와 ax 값을 리턴해준다. 

train['Survived'][train['Sex'] == 'male'].value_counts().plot.pie(
    ax=ax[0], explode=[0,0.2], autopct='%.1f%%', shadow=True)
train['Survived'][train['Sex'] == 'female'].value_counts().plot.pie(
    ax=ax[1], explode=[0,0.2], autopct='%.1f%%', shadow=True)
# 열 하나에 대해서만 나타낸 것11
# train[train['Sex']=='male']
# explode : 중심점으로부터 얼마나 떨어지게 할거냐
plt.show()
```
