```python
import tensorflow.compat.v1 as tf
tf.disable_v2_behavior()
```

```python
#1. 다중선형회귀모델
```

```python
tf.set_random_seed(777)
x1_data = [73., 93., 89., 96., 73.] # 입력 데이터
x2_data = [80., 88., 91., 98., 66.] # 모의고사라 생각 
x3_data = [75., 93., 90., 100., 70.]

y_data = [152., 185., 180., 196., 142.] # 출력 데이터 # 수능점수라 생각

x1 = tf.placeholder(tf.float32)
x2 = tf.placeholder(tf.float32)
x3 = tf.placeholder(tf.float32)

Y = tf.placeholder(tf.float32)

w1 = tf.Variable(tf.random_normal([1]), name='weight1')
w2 = tf.Variable(tf.random_normal([1]), name='weight2')
w3 = tf.Variable(tf.random_normal([1]), name='weight3')
b = tf.Variable(tf.random_normal([1]), name='bias')
```

```python
hx=w1*x1 + w2*x2 + w3*x3 + b
```

```python
cost=tf.reduce_mean( tf.square(hx-Y) ) # square = 제곱
```

```python
opt=tf.train.GradientDescentOptimizer(learning_rate=1e-5)
```

```python
train=opt.minimize(cost)
```

```python
sess = tf.Session()
sess.run(tf.global_variables_initializer())
```

```python
for step in range(2001):
    _, cv, hv = sess.run([train, cost, hx],
     feed_dict={x1:x1_data, x2:x2_data, x3:x3_data, Y:y_data})
    if step%100==0:
        print(step, 'cost:',cv, 'prediction:', hv)
        # 정답 [149.50464 186.11029 179.94537 194.88257 144.90878]
```

```python
print(sess.run([w1,w2,w3,b]))
# 회귀모델 = 1.4419925 * x1 + 0.20719662 * x2 + 0.33928198 * x3 + 2.2179823
```

```python
# 5*3
x_data = [[73., 80., 75.],
          [93., 88., 93.],
          [89., 91., 90.],
          [96., 98., 100.],
          [73., 66., 70.]]
# 5*1          
y_data = [[152.],
          [185.],
          [180.],
          [196.],
          [142.]]

# x1 = tf.placeholder(tf.float32)
# x2 = tf.placeholder(tf.float32)
# x3 = tf.placeholder(tf.float32)
X = tf.placeholder(tf.float32, shape=[None, 3]) 
# [5,3] None : 입력 데이터 건수가 정해지지 않음, 텐서플로우가 새어서 값을 정해줘

# Y = tf.placeholder(tf.float32)
Y = tf.placeholder(tf.float32, shape=[None, 1])

# w1 = tf.Variable(tf.random_normal([1]), name='weight1')
# w2 = tf.Variable(tf.random_normal([1]), name='weight2')
# w3 = tf.Variable(tf.random_normal([1]), name='weight3')
W = tf.Variable(tf.random_normal([3, 1]), name='weight')
# x * w
# [5,3] * [3,1]

b = tf.Variable(tf.random_normal([1]), name='bias')

# hx=w1*x1 + w2*x2 + w3*x3 + b
hx = tf.matmul(X, W) + b
```

```python
# 미분을 해서 cost를 낮추는 쪽으로 w와 b를 계속 업데이트 한다.

# entropy는 복잡도를 의미

# 회귀모델 -> 연속형 예측 -> 분류문제 활용하기 위해서 (by sigmoid)
# wx + b -> -inif ~ +inif -> 0 ~ 1 

# Multinoimial Classfication

# softmax function : 값을 확률로 변환해주는 함수
```

```python
from tqdm import tqdm_notebook
```

```python
# 단일 분류기 / 다중 분류기

x_data = [[1, 2],
          [2, 3],
          [3, 1],
          [4, 3],
          [5, 3],
          [6, 2]]
y_data = [[0],
          [0],
          [0],
          [1],
          [1],
          [1]]

X = tf.placeholder(tf.float32, shape=[None, 2]) #2차원 
Y = tf.placeholder(tf.float32, shape=[None, 1])
W = tf.Variable(tf.random_normal([2, 1]), name='weight')
#  x    *    w
#[6,2]  *  [2,1]
b = tf.Variable(tf.random_normal([1]), name='bias')
#hx= sigmoid(w1*x1 + w2*x2 + b)
hx=tf.sigmoid(tf.matmul(X,W) + b)  # 모델

cost = -tf.reduce_mean((Y*tf.log(hx)+(1-Y)*tf.log(1-hx)))

train = tf.train.GradientDescentOptimizer(learning_rate=0.01).minimize(cost)

pred = tf.cast(hx>0.5, dtype=tf.float32)  # tf.cast : casting 해줌 0.5보다 크면 True, 아니면 False
# dtype을 float으로 해주면 0.5보다 크면 1, 아니면 0
acc = tf.reduce_mean(tf.cast(tf.equal(pred, Y), dtype=tf.float32)) # equal : 같으면 true, 다르면 false

sess = tf.Session()

sess.run(tf.global_variables_initializer())

for step in tqdm_notebook(range(10001)):
    cv, _ = sess.run([cost, train], feed_dict={X:x_data, Y:y_data})
    if step % 200 ==0:
        print(step, cv)
```
