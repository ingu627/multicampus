```python
dataset=np.loadtxt('f:/data/pima-indians-diabetes.csv', delimiter=',')
dataset
```

```python
x=dataset[:, 0:8]
y=dataset[:,8]
```

```python
x.shape # (768, 8)
y.shape # (768,)
```

```python
# 정규화
from sklearn.preprocessing import MinMaxScaler
x_scaled = MinMaxScaler().fit_transform(x)
x_scaled

# 모델 설정
model = Sequential()
model.add(Dense(units=64, input_dim=8, activation='relu'))
model.add(Dense(units=32, activation='relu'))
model.add(Dense(1, activation='sigmoid'))

# 모델 컴파일
model.compile(loss='BinaryCrossentropy', optimizer='adam', metrics=['accuracy'])

# 모델 실행 
hist = model.fit(x_scaled, y, epochs=1000, batch_size=5, verbose=2)

# 정확도 출력
print("가장 높은 정확도 : {:.2f}%".format(sorted(hist.history["accuracy"],reverse=True)[0]))
```

```python
dataset=np.loadtxt('f:/data/pima-indians-diabetes.csv', delimiter=',')
```

```python
xtrain=dataset[0:600, :8]
ytrain=dataset[0:600, 8]
xtest=dataset[600:, :8]
ytest=dataset[600:,8]
```

```python
xtrain.shape
```

```python
xvalid = xtrain[500:, :8]
xvalid.shape
```

```python
yvalid=ytrain[500:]
yvalid.shape
```

```python
xtrain=xtrain[0:500, :8]
ytrain=ytrain[0:500]
```

```python
# model.fit()설정
# xtrain, ytrain 데이터로 학습 -> xvalid, yvalid로 검증 , metrics=accuracy
# => 모델 튜닝
# => xtest, ytest 데이터로 테스트 진행 -> 정확도 출력
```

```python
# 모델설정
model=Sequential()

model.add(Dense(64, input_dim=8, activation='relu'))

model.add(Dense(32, activation='relu'))
model.add(Dense(16, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
# 모델 최적화 함수 등 정의
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

hist=model.fit(
    xtrain, ytrain,
    epochs=600,
    batch_size=2,
    validation_data=(xvalid, yvalid), verbose=1)

print(max(hist.history['val_accuracy']))
print(max(hist.history['accuracy']))
```

```python
from sklearn.preprocessing import StandardScaler

xvalid=StandardScaler().fit_transform(xvalid)
xtrain=StandardScaler().fit_transform(xtrain)
model=Sequential()
model.add(Dense(64, input_dim=8, activation='relu'))
model.add(Dense(32, activation='relu'))
model.add(Dense(16, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
hist = model.fit(xtrain, ytrain, epochs = 300, batch_size=10, validation_data=(xvalid,yvalid))
plt.plot(hist.history['accuracy'])
```

```python
fig, loss_ax = plt.subplots()
acc_ax = loss_ax.twinx()
loss_ax.plot(hist.history['loss'], 'y', label='train loss')
loss_ax.plot(hist.history['val_loss'], 'r', label='val loss')
acc_ax.plot(hist.history['accuracy'], 'b', label='train acc')
acc_ax.plot(hist.history['val_accuracy'], 'g', label='val acc')

loss_ax.set_xlabel('epoch')
loss_ax.set_ylabel('loss')
acc_ax.set_ylabel('accuray')

loss_ax.legend(loc='upper left')
acc_ax.legend(loc='lower left')
```

```python
le=LabelEncoder()
le.fit(y)
y=le.transform(y)
y
```

```python
y=tf.keras.utils.to_categorical(y)
```

```python
# fit_transform은 트레이닝 데이터에 사용
# transform은 테스트 데이터에 사용

# 표준화 과정 (StandardScaler클래스)
# - train, test 데이터 모두 다 표준화를 해야 함

# fit_transform 함수
# : 트레이닝 데이터로 학습된 스케일러를 통해 테스트 데이터를 스케일링 하겠다

# transform 함수
# : 테스트 데이터로 학습된 스케일러를 통해 테스트 데이터를 스케일링 하겠다
```

```python
# 아이리스 학습 데이터로 MinMaxScaler() 생성 -> fit() -> trnasform() -> 분류기 학습
```

```python
# scaler.fit(xtrain)
# scaled_xtrain=scaler.transform(xtrain)
# => scaled_xtrain = scaler.fit_transform(xtrain)

# 단 테스트데이터에 대해 아래와 같이 적용해서는 안됨
# scaler가 기존 학습 데이터에 fit했던 기준을 모두 무시하고 
# 다시 테스트 데이터를 기반으로 기준을 적용
# sclaed_xtest = scaler.fit_transform(xtest)

# 트레인, 테스트 데이터를 나누기 전에 스케일링 등 데이터 전처리를 미리 해주는 게 좋다
# fit () -> transform 또는 fit_transform() 모두 사용 가능
```

```python
scaler = StandardScaler()
```

```python
#학습/테스트 데이터로 나누기 전에 fit과 transform으로 변환(fit_transform 함수도 가능)
scaler.fit(df.iloc[:,0:4])
```

```python
scaled_df=scaler.transform(df.iloc[:,0:4])
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(
    scaled_df, y, test_size=0.3, random_state=1022
)
```

```python
print(xtrain.shape)
print(ytrain.shape)
print(xtest.shape)
print(ytest.shape)
```

```python
model=Sequential()
```

```python
model.add(Dense(16, input_dim=4, activation='relu'))
model.add(Dense(3, activation='softmax'))
```

```python
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
model.fit(xtrain, ytrain, epochs=100, batch_size=1)
```

```python
yhat=model.predict(xtest)
```

```python
np.sum(yhat[0])
# np.argmax(yhat[1])
```

```python
ytest
```

```python
#yhat과 ytest 비교하여 정확도 출력
sum=0
for i in range(len(ytest)):
    if np.argmax(yhat[i]) == np.argmax(ytest[i]):
        sum+=1

acc=sum/len(ytest)*100
print(acc)
```
